```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
```

## Gráficación de regresión lineal.

```python
%matplotlib inline 
a = 0
b = 1
n = 200
x = np.linspace(0,n,n)
y = a + b*x +20*np.random.randn(n)
plt.plot(x,y,'.')
plt.plot(a+b*x, '-r')
plt.hlines(np.mean(y), xmin=x[0], xmax=x[n-1])   #valor promedio de la y.
```

```python
ybarra = np.mean(y)
xbarra = np.mean(x)
numerador= sum((x-xbarra)*(y-ybarra))
numerador
denom1 = np.sqrt(sum((x-xbarra)**2))
denom2 = np.sqrt(sum((y-ybarra)**2))

denom = denom1*denom2
numerador/denom
```
